# Set-up the project environment for Formula1 Project
1. Create External Location databricks-course-ext-dl1-formula1-incr
1. Create Catalog formula1_incr
1. Create Schemas landing, bronze, silver and gold
1. Create Volume Files in the landing schema

### Access Cloud Storage

```sql
%fs ls 'abfss://formula1-incr@datbricksstrgaccextdl.dfs.core.windows.net/landing'
```

### Create External Location

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS formula1_incrload_ext
URL 'abfss://formula1-incr@datbricksstrgaccextdl.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `formula1_storage_creds`)
COMMENT 'External location for the formula1-incr container';
```

### Create Catalog formula1

```sql
SHOW CATALOGS;
```

```sql
CREATE CATALOG IF NOT EXISTS formula1_incr
   MANAGED LOCATION 'abfss://formula1-incr@datbricksstrgaccextdl.dfs.core.windows.net/' 
   COMMENT 'This is the main catalog for the formula1 project' ;
```

### Create Schemas landing, bronze, silver, gold

```sql
CREATE SCHEMA IF NOT EXISTS formula1_incr.landing;
CREATE SCHEMA IF NOT EXISTS formula1_incr.bronze
    MANAGED LOCATION 'abfss://formula1-incr@datbricksstrgaccextdl.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS formula1_incr.silver
    MANAGED LOCATION 'abfss://formula1-incr@datbricksstrgaccextdl.dfs.core.windows.net/silver';
CREATE SCHEMA IF NOT EXISTS formula1_incr.gold
    MANAGED LOCATION 'abfss://formula1-incr@datbricksstrgaccextdl.dfs.core.windows.net/gold';         
```

### Create Volume Files

```sql
CREATE EXTERNAL VOLUME formula1_incr.landing.files
LOCATION 'abfss://formula1-incr@datbricksstrgaccextdl.dfs.core.windows.net/landing';
```
